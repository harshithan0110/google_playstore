```python
# --------------------------------------------------------
# GOOGLE PLAY STORE APPS DATA ANALYSIS PROJECT
# Tools: Python, Pandas, Matplotlib, Seaborn
# --------------------------------------------------------

# Step 0: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Step 0.1: Ignore unnecessary warnings
warnings.filterwarnings("ignore")

# --------------------------------------------------------
# Step 1: Load dataset
# --------------------------------------------------------
# Make sure 'googleplaystore.csv' is in the same folder as this notebook
df = pd.read_csv(r"C:\Users\Harshitha N\hash\googleplaystore.csv")

# --------------------------------------------------------
# Step 2: Display basic info
# --------------------------------------------------------
print("Dataset Info:")
print(df.info())
print("\nDataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())

# --------------------------------------------------------
# Step 3: Data Cleaning
# --------------------------------------------------------

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Handle missing values
df.dropna(inplace=True)

# Clean 'Installs' column (remove + and , then convert to int)
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
df['Installs'] = df['Installs'].astype(int)

# Clean 'Price' column (remove $ and convert to float)
df['Price'] = df['Price'].str.replace('$', '', regex=False).astype(float)

# Convert 'Reviews' to int
df['Reviews'] = df['Reviews'].astype(int)

# --------------------------------------------------------
# Step 4: Exploratory Data Analysis (EDA)
# --------------------------------------------------------

print("\nTop 10 Categories by Number of Apps:")
print(df['Category'].value_counts().head(10))

# Visualization 1: Top 10 App Categories
plt.figure(figsize=(12,6))
df['Category'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 App Categories')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.show()

# Visualization 2: Rating Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Rating'], bins=20, kde=True, color='orange')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Visualization 3: Relationship between Reviews and Ratings
plt.figure(figsize=(10,6))
sns.scatterplot(x='Reviews', y='Rating', data=df, alpha=0.6)
plt.title('Reviews vs Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('App Rating')
plt.show()

# Visualization 4: Free vs Paid Apps
plt.figure(figsize=(6,6))
df['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen','pink'])
plt.title('Free vs Paid Apps')
plt.ylabel('')
plt.show()

# Visualization 5: Average Rating by Category
plt.figure(figsize=(12,6))
avg_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_rating.index, y=avg_rating.values, hue=avg_rating.index, palette='coolwarm', legend=False)
plt.title('Top 10 Categories with Highest Average Ratings')
plt.xticks(rotation=45)
plt.ylabel('Average Rating')
plt.xlabel('Category')
plt.show()

# --------------------------------------------------------
# Step 5: Insights / Conclusions
# --------------------------------------------------------
print("\n KEY INSIGHTS:")
print("1 Most apps on Google Play Store are free.")
print("2 The 'GAME' and 'TOOLS' categories have the highest number of apps.")
print("3 Paid apps tend to have slightly higher average ratings than free apps.")
print("4 Higher reviews generally indicate higher installs  showing user engagement.")
print("5 Ratings are mostly concentrated between 4.0 and 4.5  showing good overall quality.")
```
